```python
def IQR(x):
    q1 = x.quantile(0.25)
    q3 = x.quantile(0.75)
    iqr = q3 - q1
    LF = q1 - (1.5*iqr)
    UF = q3 + (1.5*iqr)
    print("LF-->",LF)
    print("UF-->",UF)
```
